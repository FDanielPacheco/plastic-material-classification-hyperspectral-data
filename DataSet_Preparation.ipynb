```python
import numpy as np 
import pandas as pd
 
MetaData = pd.read_csv("MetaData.csv")
Data = pd.read_csv("Data.csv")

Col_Plastic_Type = MetaData.loc[:, 'Polymer identified via microFTIR']
Col_Sample_Code = MetaData.loc[:,'Sample code']
N_Samples = Data.shape[1]
 

# Read data
Data = pd.read_csv("Data.csv")
MetaData = pd.read_csv("MetaData.csv")
wavelengths = Data.iloc[:, 0].astype(float).values
data = Data.iloc[:, 1:].astype(float).values

sample_codes = MetaData.iloc[:, 0].astype(str).values
replicates = MetaData.iloc[:, 16].astype(int).values

# Samples to exclude

EXCLUDED_SAMPLES = {
    "White_reference ",
    "Water_16_sediment",
    "Clear water",
    "Background_lid",
    "Black_cloth_calibrationFacility",
}
```

```python

# Process data
new_data = []
new_sample_codes = []
sample_ranges = []

current_sum = None
num_replicates = 0
range_start = 0

num_samples = len(sample_codes)

for i in range(num_samples - 1):

    if sample_codes[i] in EXCLUDED_SAMPLES:
        continue

    if replicates[i] == 1:
        current_sum = data[:, i].copy()
        num_replicates = 1
        range_start = i
    else:
        current_sum += data[:, i]
        num_replicates += 1

    # End of replicate block
    if replicates[i + 1] != replicates[i] + 1:
        mean_spectrum = current_sum / num_replicates

        new_data.append(mean_spectrum)
        new_sample_codes.append(sample_codes[i])
        sample_ranges.append(f"{range_start}-{i}")
# ------------------------------------------------------------------
# Export to CSV
# ------------------------------------------------------------------
df_export = pd.DataFrame(
    np.array(new_data),
    index=new_sample_codes,
    columns=wavelengths
)

df_export.insert(0, "Sample_Range", sample_ranges)
df_export.to_csv("Data_AVG.csv")
```
