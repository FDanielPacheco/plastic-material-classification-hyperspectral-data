### Description

This notebook is used for the following purposes:
- Measure:
  - True risk with the probability of missclassification 
    - $P( f(\vec{\text{x}}) \neq y )$
  - Empirical risk the proportion of misclassified examples on training data
  - Quality measurment with Misses and False alarms
    - False Positive ($FP$) rate: $FP / (FP + TN)$
    - False Negative ($FN$) rate: $FN / (FN + TP)$
    - True Positive ($TP$) rate: $TP / (TP + FN)$
    - Precision: $TP / (TP + FP) $ 

    - Really $C$ in rows, Predicted $C$ in columns:
      |      | Yes  | No   |
      | ---- | ---- | ---- |
      | Yes  |  TP  | FN   |
      |  No  | FP   | TN   |
